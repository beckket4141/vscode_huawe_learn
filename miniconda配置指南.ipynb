# Miniconda 配置指南

本指南将详细介绍如何将 Miniconda 添加到系统环境变量中，以便在终端中直接使用 conda 命令。

## 环境信息

根据您提供的信息，您的 Miniconda 安装路径为：`C:\\Users\\86136\\miniconda3`

## Windows 系统配置步骤

### 方法一：通过图形界面添加环境变量

1. 右键点击“此电脑”或“我的电脑”，选择“属性"
2. 点击左侧的“高级系统设置"
3. 在弹出的系统属性窗口中，点击“环境变量”按钮
4. 在“系统变量”区域找到并选中`Path`变量，点击“编辑”
5. 点击“新建”，添加以下路径：
   - `C:\Users\86136\miniconda3`
   - `C:\Users\86136\miniconda3\Scripts`
   - `C:\Users\86136\miniconda3\Library\bin`
6. 点击“确定”保存所有更改
7. 关闭所有已打开的命令提示符窗口，重新打开一个新的命令提示符窗口
8. 输入 `conda --version` 验证配置是否成功

### 方法二：通过命令行自动配置

您也可以通过以下命令将 Miniconda 添加到环境变量中：

1. 打开命令提示符（以管理员身份运行）
2. 执行以下命令：

```cmd
setx PATH "%PATH%;C:\Users\86136\miniconda3;C:\Users\86136\miniconda3\Scripts;C:\Users\86136\miniconda3\Library\bin"
```

3. 关闭命令提示符窗口，重新打开一个新的命令提示符窗口
4. 输入 `conda --version` 验证配置是否成功

## 验证配置

配置完成后，您可以使用以下命令验证是否配置成功：

```bash
# 查看conda版本
conda --version

# 查看conda信息
conda info

# 查看已安装的包
conda list
```

## 常用 Conda 命令

配置完成后，您可以使用以下常用 conda 命令：

```bash
# 创建新环境
conda create -n myenv python=3.8

# 激活环境
conda activate myenv

# 退出环境
conda deactivate

# 安装包
conda install package_name

# 查看所有环境
conda env list
```

## 注意事项

1. 修改环境变量后，必须重新打开命令提示符窗口才能生效
2. 如果您使用的是 PowerShell，同样需要重新打开 PowerShell 窗口
3. 如果遇到权限问题，请尝试以管理员身份运行命令提示符
4. 不同用户的 Miniconda 安装路径可能不同，请根据实际情况调整路径

完成以上步骤后，您就可以在任何终端窗口中使用 conda 命令了。
